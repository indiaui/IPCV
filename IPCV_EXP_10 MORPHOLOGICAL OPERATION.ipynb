```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
closing_img = cv2.imread('/content/closing.png')
print("Closing Image")
cv2_imshow(closing_img)

opening_img = cv2.imread('/content/opening.png')
print("Opening Image")
cv2_imshow(opening_img)

dilation_img = cv2.imread('/content/dilation.png')
print("Dilation Image")
cv2_imshow(dilation_img)

erosion_img = cv2.imread('/content/erosion.png')
print("Erosion Image")
cv2_imshow(erosion_img)
```

```python
gray_opening = cv2.cvtColor(opening_img, cv2.COLOR_BGR2GRAY)
print("Grayscale Opening Image")
cv2_imshow(gray_opening)

gray_closing = cv2.cvtColor(closing_img, cv2.COLOR_BGR2GRAY)
print("Grayscale Closing Image")
cv2_imshow(gray_closing)

gray_dilation = cv2.cvtColor(dilation_img, cv2.COLOR_BGR2GRAY)
print("Grayscale Dilation Image")
cv2_imshow(gray_dilation)

gray_erosion = cv2.cvtColor(erosion_img, cv2.COLOR_BGR2GRAY)
print("Grayscale Erosion Image")
cv2_imshow(gray_erosion)
```

```python
# Step 2: Convert to Binary (thresholding)
_, binary_open = cv2.threshold(gray_opening, 127, 255, cv2.THRESH_BINARY)
print("Binary Opening Image")
cv2_imshow(binary_open)

_, binary_close = cv2.threshold(gray_closing, 127, 255, cv2.THRESH_BINARY)
print("Binary Closing Image")
cv2_imshow(binary_close)

_, binary_dilate = cv2.threshold(gray_dilation, 127, 255, cv2.THRESH_BINARY)
print("Binary Dilation Image")
cv2_imshow(binary_dilate)

_, binary_erode = cv2.threshold(gray_erosion, 127, 255, cv2.THRESH_BINARY)
print("Binary Erosion Image")
cv2_imshow(binary_erode)
```

```python
#Create Kernel (matrix for morphological operations)
kernel = np.ones((5,5), np.uint8)
```

```python
erosion = cv2.erode(binary_erode, kernel, iterations=1)
print("Before Erosion Image")
cv2_imshow(binary_erode)
print("After Erosion Image")
cv2_imshow(erosion)
```

```python
dilation = cv2.dilate(binary_dilate, kernel, iterations=1)
print("Before Dilation Image")
cv2_imshow(binary_dilate)
print("After Dilation Image")
cv2_imshow(dilation)
```

```python
opening = cv2.morphologyEx(binary_open, cv2.MORPH_OPEN, kernel)
print("Before Opening Image")
cv2_imshow(binary_open)
print("After Opening Image")
cv2_imshow(opening)
```

```python
closing = cv2.morphologyEx(binary_close, cv2.MORPH_CLOSE, kernel)
print("Before Closing Image")
cv2_imshow(binary_close)
print("After Closing Image")
cv2_imshow(closing)

```
