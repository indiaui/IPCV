```python
import cv2
import numpy as np
from google.colab import files
from google.colab.patches import cv2_imshow
```

```python
#Reading an Image
import cv2
img = cv2.imread(r"/content/IMG.jpg",0)
```

```python
#Display an Image - using Google Colab
from google.colab.patches import cv2_imshow
cv2_imshow(img)
```

```python
# Apply Gaussian Blur with same kernel size but different sigma

k = 15      # kernel size
s1 = 1.0    # sigma 1
s2 = 4.0    # sigma 2

g1 = cv2.GaussianBlur(img, (k, k), sigmaX=s1)
cv2_imshow(g1)
g2 = cv2.GaussianBlur(img, (k, k), sigmaX=s2)
cv2_imshow(g2)
# Compute Difference of Gaussians
dog1 = cv2.subtract(g1, g2)
dog1 = cv2.normalize(dog1, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)
cv2_imshow(dog1)
```

```python
# Apply Gaussian Blur with different kernel sizes but same sigma

k1 = 7
k2 = 25
s = 2.0

g3 = cv2.GaussianBlur(img, (k1, k1), sigmaX=s)
cv2_imshow(g3)
g4 = cv2.GaussianBlur(img, (k2, k2), sigmaX=s)
cv2_imshow(g4)
dog2 = cv2.subtract(g3, g4)
dog2 = cv2.normalize(dog2, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)
cv2_imshow(dog2)
```
