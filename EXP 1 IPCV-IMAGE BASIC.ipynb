```python
#Read Image
import cv2
img = cv2.imread(r"D:\COLLEGE - SEM 5\IMAGE PROCESSING AND COMPUTER VISION\EXP 1\champaklal.jpg",1)
```

```python
#Display Image
cv2.imshow("Image ",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Dimensions get
row,col,channels = img.shape
print("Row:", row)
print("Col:", col)
print("Channels:", channels)
```

```python
# Data type of image
img.dtype
```

```python
#Splitting into 3 colour layers- RGB channels
b_channel, g_channel, r_channel = cv2.split(img)
cv2.imshow('Blue Channel', b_channel)
cv2.imshow('Green Channel', g_channel)
cv2.imshow('Red Channel', r_channel)
cv2.waitKey(0) # Wait indefinitely until a key is pressed
cv2.destroyAllWindows() # Close all OpenCV windows
```

```python
##Splitting into 3 colour layers- RGB channels and giving colors to rgb layers
import cv2
import numpy as np
b, g, r = cv2.split(img)
zeros = np.zeros(img.shape[:2], dtype="uint8")
blue_channel_colored = cv2.merge([b, zeros, zeros])
cv2.imshow("Blue Channel", blue_channel_colored)
green_channel_colored = cv2.merge([zeros, g, zeros])
cv2.imshow("Green Channel", green_channel_colored)
red_channel_colored = cv2.merge([zeros, zeros, r])
cv2.imshow("Red Channel", red_channel_colored)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
img01 = cv2.imread(r"champaklal.jpg",1)
img02 = cv2.imread(r"doraemon.png",1)
```

```python
cv2.imshow("Image ",img01)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imshow("Image ",img02)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Resizing image 2 matching to image 1
img02_resized = cv2.resize(img02,(img01.shape[1],img01.shape[0]))
```

```python
#Adding two images
cv2.add(blue_channel_colored, green_channel_colored)
```

```python
#Adding two images
cv2.add(img01,img02_resized)
```

```python
#Arithmetic operation - Adding two images
added_image = cv2.add(blue_channel_colored, green_channel_colored)
cv2.imshow("Added Image",added_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Arithmetic operation - Adding two images
added_image = cv2.add(img01,img02_resized)
cv2.imshow("Added Image",added_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Arithmetic operation - Subtracting Images
subtract_image = cv2.subtract(img01,img02_resized)
cv2.imshow("Subtract Image",subtract_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Arithmetic operation - Subtracting Images
subtract_image = cv2.subtract(blue_channel_colored, green_channel_colored)
cv2.imshow("Subtract Image",subtract_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Arithmetic Operation - Image Alpha Blending - Weighted Average
# Perform alpha blending using addWeighted
# Arguments: src1, alpha, src2, beta (1-alpha), gamma (scalar added to each sum)
blend = cv2.addWeighted(img01, 0.7, img02_resized, 0.3, -10)
cv2.imshow("Blend Image",blend)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Logical Operation - AND operation
And_image = cv2.bitwise_and(img01,img02_resized)
cv2.imshow("AND Image",And_image )
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Logical Operation - OR operation
OR_image = cv2.bitwise_or(img01,img02_resized)
cv2.imshow("OR Image",OR_image )
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Logical Operation - NOT operation
NOT_image = cv2.bitwise_not(img01)
cv2.imshow("NOT Image",NOT_image )
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Logical Operation - NOT operation
NOT_image = cv2.bitwise_not(img02_resized)
cv2.imshow("NOT Image",NOT_image )
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Logical Operation - NOT operation
NOT_image = cv2.bitwise_not(red_channel_colored)
cv2.imshow("NOT Image",NOT_image )
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Logical Operation - XOR operation
XOR_image = cv2.bitwise_xor(img01,img02_resized)
cv2.imshow("XOR Image",XOR_image )
cv2.waitKey(0)
cv2.destroyAllWindows()
```
