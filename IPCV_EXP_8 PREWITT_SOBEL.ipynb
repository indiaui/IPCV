# ***Prewitt Mask - 1st Approach***

```python
#Reading an Image
import cv2
img = cv2.imread(r"/content/horizontal_vertical_lines.png",0)
```

```python
#Display an Image - using Google Colab
from google.colab.patches import cv2_imshow
cv2_imshow(img)
```

```python
# Obtain number of rows and columns
row,col = img.shape
```

```python
print("Rows of image is : ",row)
print("Columns of image is : ",col)
```

```python
#Prewitt Mask - Fx
mask_fx = np.array([[-1,-1,-1],[0,0,0],[1,1,1]])
print("Fx mask is : \n",mask_fx)
```

```python
#Prewitt Mask - Fy
mask_fy = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
print("Fy mask is : \n",mask_fy)
```

```python
# Convolve the 3X3 mask over the image -> Fx Mask
fx_image = np.zeros([row, col])

for i in range(1,row-1):
  for j in range(1,col-1):
    temp = img[i-1, j-1]*mask_fx[0, 0]+img[i-1, j]*mask_fx[0, 1]+img[i-1, j + 1]*mask_fx[0, 2]+img[i, j-1]*mask_fx[1, 0]+ img[i, j]*mask_fx[1, 1]+img[i, j + 1]*mask_fx[1, 2]+img[i + 1, j-1]*mask_fx[2, 0]+img[i + 1, j]*mask_fx[2, 1]+img[i + 1, j + 1]*mask_fx[2, 2]
    fx_image[i, j]= temp

print("Original image is : ")
cv2_imshow(img)
print("Fx image is : ")
cv2_imshow(fx_image)
```

```python
# Convolve the 3X3 mask over the image -> Fy Mask
fy_image = np.zeros([row, col])

for i in range(1,row-1):
  for j in range(1,col-1):
    temp = img[i-1, j-1]*mask_fy[0, 0]+img[i-1, j]*mask_fy[0, 1]+img[i-1, j + 1]*mask_fy[0, 2]+img[i, j-1]*mask_fy[1, 0]+ img[i, j]*mask_fy[1, 1]+img[i, j + 1]*mask_fy[1, 2]+img[i + 1, j-1]*mask_fy[2, 0]+img[i + 1, j]*mask_fy[2, 1]+img[i + 1, j + 1]*mask_fy[2, 2]
    fy_image[i, j]= temp

print("Original image is : ")
cv2_imshow(img)
print("Fy image is : ")
cv2_imshow(fy_image)
```

```python
#Combination of Image - Fx+Fy
image = np.zeros([row, col])

for i in range(1,row-1):
  for j in range(1,col-1):
    temp = abs(fx_image[i, j])+abs(fy_image[i, j])
    image[i, j]= temp

print("Original image is : ")
cv2_imshow(img)
print("Prewitt Final image is : ")
cv2_imshow(image)
```
