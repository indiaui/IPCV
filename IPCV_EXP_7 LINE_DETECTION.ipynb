# ***Horizontal Line Detection***

```python
#Reading an Image
import cv2
img = cv2.imread(r"/content/horizontal_vertical_lines.png",0)
```

```python
#Display an Image - using Google Colab
from google.colab.patches import cv2_imshow
cv2_imshow(img)
```

```python
# Obtain number of rows and columns
row,col = img.shape
```

```python
print("Rows of image is : ",row)
print("Columns of image is : ",col)
```

```python
#Horizontal Mask
mask = np.array([[-1,-1,-1],[2,2,2],[-1,-1,-1]])
print("Horizontal mask is : \n",mask)
```

```python
# Convolve the 3X3 mask over the image -> Horizontal Mask
detected_image = np.zeros([row, col])

for i in range(1,row-1):
  for j in range(1,col-1):
    temp = img[i-1, j-1]*mask[0, 0]+img[i-1, j]*mask[0, 1]+img[i-1, j + 1]*mask[0, 2]+img[i, j-1]*mask[1, 0]+ img[i, j]*mask[1, 1]+img[i, j + 1]*mask[1, 2]+img[i + 1, j-1]*mask[2, 0]+img[i + 1, j]*mask[2, 1]+img[i + 1, j + 1]*mask[2, 2]
    detected_image[i, j]= temp

print("Original image is : ")
cv2_imshow(img)
print("Horizontal Line detected image is : ")
cv2_imshow(detected_image)
```
